# Data Exploration - Crop Disease Dataset
This notebook explores the PlantVillage dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import cv2
import numpy as np
import seaborn as sns
```

```python
# Load metadata
df = pd.read_csv('../data/processed/dataset_metadata.csv')
print(f"Total images: {len(df)}")
print(f"Classes: {df['class_label'].nunique()}")
df.head()
```

```python
# Class distribution
plt.figure(figsize=(12, 6))
df['class_label'].value_counts().plot(kind='bar')
plt.title('Images per Class')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Show sample images
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
for i, ax in enumerate(axes.flat):
    img_path = df.iloc[i * 1000]['image_path']
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    ax.imshow(img)
    ax.set_title(df.iloc[i * 1000]['class_label'])
    ax.axis('off')
plt.tight_layout()
plt.show()
```

```python
# Healthy vs Diseased
plt.figure(figsize=(8, 6))
df['is_healthy'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Healthy vs Diseased Distribution')
plt.ylabel('')
plt.show()
```

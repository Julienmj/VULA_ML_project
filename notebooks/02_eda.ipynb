# Exploratory Data Analysis - Crop Disease Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
%matplotlib inline

CSV_PATH = r'c:\Users\Administrator\Desktop\AIC\data\processed\dataset_metadata.csv'
```

## 1. Load Data

```python
df = pd.read_csv(CSV_PATH)
print(f"Dataset shape: {df.shape}")
df.head()
```

## 2. Class Distribution

```python
print(f"Total Classes: {df['class_label'].nunique()}")
print(f"Total Crops: {df['crop'].nunique()}")

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
df['crop'].value_counts().plot(kind='bar', ax=axes[0], color='steelblue')
axes[0].set_title('Samples per Crop')
df['is_healthy'].value_counts().plot(kind='bar', ax=axes[1], color=['red', 'green'])
axes[1].set_title('Healthy vs Diseased')
plt.tight_layout()
plt.show()
```

## 3. Summary

```python
print(f'Total Images: {len(df):,}')
print(f'Crops: {df["crop"].nunique()}')
print(f'Diseases: {df["disease"].nunique()}')
print(f'Healthy: {df["is_healthy"].sum():,}')
print(f'Diseased: {(~df["is_healthy"]).sum():,}')
```
